```python
#Loading excel file with scraped Title and Articcle
import pandas as pd
df=pd.read_excel(r"C:\Users\Lenovo\Desktop\Article.xlsx")
df.head()
```

```python
#Creating a list of stopwords from stopwords.txt which contains all the stopwords txt file
mylist=[]
with open(r'C:\Users\Lenovo\Desktop\StopWords.txt', 'r') as f:
    f2=f.readlines()
    for line in f2:
        mylist.append(line)
```

```python
mylist
```

```python
mylist = [item.strip() for item in mylist]
```

```python
mylist
```

```python
# Text cleaning and removing punctuations
import re
def clean_data(Content):
    Content = re.sub(r'[^a-zA-Z.]', ' ',Content)
    Content = Content.lower()
    return Content
```

```python
df['Article'] = df['Article'].apply(clean_data)
```

```python
# Removing stopwrods
def remove_stop_words(Content):
    Content_minus_sw = []
    
    z=mylist 
    Content = Content.split()
    Content = [Content_minus_sw.append(word) for word in Content if word not in z]
    Content = ' '.join(Content_minus_sw)
    return Content
```

```python
df['Article'] = df['Article'].apply(remove_stop_words)
```

```python
df['Article']
```

```python
#Creating dictionary of positive and negative wor 
pos=[]
with open(r'E:\Download\20211030 Test Assignment-20221227T043433Z-001\20211030 Test Assignment\MasterDictionary\positive-words.txt', 'r') as f:
    f2=f.readlines()
    for line in f2:
        pos.append(line)
neg=[]
with open(r'E:\Download\20211030 Test Assignment-20221227T043433Z-001\20211030 Test Assignment\MasterDictionary\negative-words.txt', 'r') as f:
    f2=f.readlines()
    for line in f2:
        neg.append(line)
        
pos = [item.strip() for item in pos]
neg = [item.strip() for item in neg]

mydict={"Positive":pos, "Negative":neg}
```

```python
mydict
```

```python
df.info()
```

```python
df["Article"]=df["Article"].astype("string")
```

```python
#Counting positive and negative words in each Article
posi=[]
negi=[]
for x in df["Article"]:
    pos_count=0
    neg_count=0
    words=x.split()
    wordcount=len(words)
    for i in range(0, wordcount):
        if words[i] in pos:
            pos_count += 1
        
        
        if words[i] in neg:
            neg_count += 1
    posi.append(pos_count)
    negi.append(neg_count)
```

```python
#adding column of positive_score and negative_score
df["Positive_score"]=posi
df["Negative_score"]=negi
```

```python
#calculating polarity score 
df["Polarity_score"]= (df["Positive_score"] - df["Negative_score"] )/((df["Positive_score"] + df["Negative_score"] )+ 0.000001)
```

```python
df["Polarity_score"]
```

```python
#calculating total words in an Article
total_words=[]
for x in df["Article"]:
    y=len(x.split())
    total_words.append(y)
    
    
```

```python
import numpy as np
word_count=np.array(total_words)
```

```python
word_count
```

```python
#adding a column with subjectivity score
df["Subjectivity_score"]= (df["Positive_score"]+df["Negative_score"]) / (word_count  +  0.000001 )
```

```python
#Finding number of sentences in An Article
num_sentences=[]

for x in df["Article"]:
    y=re.split(r'[.]+', x)
    num_sentence=len(y)
    num_sentences.append(num_sentence)
```

```python
total_sentences=np.array(num_sentences)
```

```python
df
```

```python
#calculating average sentence length
df["Average_sentence_length "]= word_count / total_sentences
```

```python
df
```

```python
#calculating number of complex words in an article
complexwords=[]
    
for text in df["Article"]:
    complexWord = 0
   
    for word in text.split():
        vowels=0
        
        if word.endswith(('es','ed')):
            pass
        else:
            for w in word:
                if(w=='a' or w=='e' or w=='i' or w=='o' or w=='u'):
                    vowels += 1
            if(vowels > 2):
                complexWord += 1
    complexwords.append(complexWord)
    
    
    
```

```python
complexwords
```

```python
Count_complexwords=np.array(complexwords)
```

```python
#Calculating complex word perctange
df["Complexwords_percent"]=Count_complexwords/ word_count
```

```python
df["Complexwords_percent"]
```

```python
#calculating fog index
df["Fog Index"]=0.4 *(df["Average_sentence_length "] + df["Complexwords_percent"])
```

```python
#calculating average word per sentence
df["Average_word_per_sentence"]=word_count/total_sentences
```

```python
# adding a column for word_count calculated earlier
df["Word_count"]=word_count
```

```python
# counting all vowels in every word in an Article
vowelcount=[]
    
for text in df["Article"]:
    
    vowels=0
   
    for word in text.split():
        
        
        if word.endswith(('es','ed')):
            pass
        else:
            for w in word:
                if(w=='a' or w=='e' or w=='i' or w=='o' or w=='u'):
                    vowels += 1
            
    vowelcount.append(vowels)
    
```

```python
#Finding pronouns in an Article
pronouns_list=[]

for x in df["Article"]:
    y=re.compile(r'\b(I|we|my|ours|(?-i:us))\b',re.I)
    pronouns = y.findall(x)
    
    pronouns_list.append(pronouns)
```
